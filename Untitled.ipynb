```python
import os
import csv 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file = ('Incident-type-and-response-times.csv')
```

```python
df = pd.read_csv(file, encoding="ISO-8859-1")
```

```python
needed_df = pd.DataFrame(df, columns = ['Basic Incident Number (FD1)','Basic Type Of Alarm (FD1.50)', 'Basic Incident Geocoded Latitude','Basic Incident Geocoded Longitude','Basic Incident First Unit Arrived Apparatus Resource ID','Basic Incident Date Time'])
needed_df = needed_df.dropna()
needed_df.head()
```

```python
stats = {}
for row in needed_df.iterrows():
    Basic_Alarm = row[1]['Basic Type Of Alarm (FD1.50)']
    value = row[1]["Basic Incident Date Time"]
    value = value.split(" ")
    value = value[0].split("/")
    month = value[0]
    year = value[2]
    day = value[1]
    if (int(year)==2019):
         yrmo = year + month 
         if (yrmo in stats):
            if (Basic_Alarm in stats[yrmo]):
              stats[yrmo][Basic_Alarm]+=1
              stats[yrmo]["total"]+=1
            else:
              stats[yrmo][Basic_Alarm]=1
              stats[yrmo]['total']=1
         else:
            stats[yrmo]={}
            stats[yrmo][Basic_Alarm]=1
            stats[yrmo]['total']=1
for yrmo in stats:
    for basical in stats[yrmo]:
        print (yrmo,basical,stats[yrmo][basical])
```
